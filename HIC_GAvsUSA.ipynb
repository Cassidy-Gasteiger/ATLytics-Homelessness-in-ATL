```python
import os
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
os.getcwd()
```

```python
#import PIT data for 2018-2022 ia a list of dataframes
xls = pd.ExcelFile('2007-2022-HIC-Counts-by-CoC (2).xlsx')

n = 2022-2018+1
HIC_df_list = [pd.DataFrame() for _ in range(n)]

for num,year in enumerate(range(2022,2018-1,-1)):
    HIC_df_list[num] = pd.read_excel(xls, str(year), skiprows=[0])
    HIC_df_list[num].insert(1, 'Year', year)
```

```python
#concatenate all dataframes into one
HIC_df = pd.DataFrame()
for df in HIC_df_list:
    try:
        df = df[df['CoC'].notna()]
        HIC_df = pd.concat([HIC_df, df])
    except:
        df = df.rename(columns={'CoC Number': 'CoC'})#some years the area code column is named CoC and some others it's named CoC Number
        df = df[df['CoC'].notna()]
        HIC_df = pd.concat([HIC_df, df])
```

```python
#Filter out total rows
HIC_df = HIC_df[HIC_df.CoC!='Total']
HIC_df[['State','Code']] = HIC_df['CoC'].str.split("-").to_list()
HIC_df
```

```python
cols_to_keep = ['State','Year', 'Dedicated Veteran Beds (ES)','Dedicated Veteran Beds (ES, TH, SH)','Dedicated Veteran Beds (SH)','Dedicated Veteran Beds (TH)','Dedicated Youth Beds (ES)','Dedicated Youth Beds (ES, TH, SH)','Dedicated Youth Beds (SH)',
                'Dedicated Youth Beds (TH)','Total Beds for Households with Children (ES)','Total Beds for Households with Children (ES, TH, SH)', 'Total Beds for Households with Children (SH)','Total Beds for Households with Children (TH)',
                'Total Beds for Households with only Children (ES)','Total Beds for Households with only Children (ES, TH, SH)','Total Beds for Households with only Children (SH)','Total Beds for Households with only Children (TH)',
                'Total Beds for Households without Children (ES)','Total Beds for Households without Children (ES, TH, SH)','Total Beds for Households without Children (SH)','Total Beds for Households without Children (TH)',
                'Total Non-DV Year-Round Beds (ES)','Total Non-DV Year-Round Beds (ES, TH, SH)','Total Non-DV Year-Round Beds (SH)','Total Non-DV Year-Round Beds (TH)','Total Overflow Beds (ES)','Total Seasonal Beds (ES)',
                'Total Units for Households with Children (ES)', 'Total Units for Households with Children (ES, TH, SH)', 'Total Units for Households with Children (SH)', 'Total Units for Households with Children (TH)', 'Total Year-Round Beds (ES)',
                'Total Year-Round Beds (ES).1', 'Total Year-Round Beds (ES, TH, SH)', 'Total Year-Round Beds (SH)', 'Total Year-Round Beds (TH)']
```

```python
new_df = HIC_df[cols_to_keep] #We're only keeping the columns we want
new_df
```

```python
new_df = new_df.groupby(['State','Year']).sum().reset_index() #Add all state area codes
melt_df = new_df.melt(id_vars=['State','Year'], var_name='Column', value_name='Values') #melt the dataframe before pivoting
melt_df
```

```python
final_df = melt_df.pivot(index=['State','Column'], columns='Year')['Values'].reset_index()#now our dataframe is in a time series format
final_df
```

```python
#Create a column for Program Type
final_df[['Column_New','Program_Type']] = final_df.Column.str.split('(').tolist()
final_df.Column_New = final_df.Column_New.str.strip()
final_df[['Program_Type','to_drop']] = final_df.Program_Type.str.split(')').tolist()
final_df = final_df.drop(['Column','to_drop'], axis=1)
final_df
```

```python
final_df.to_csv('HIC_GAvsUSA.csv', index=False)
```

```python
changes_df = final_df[(final_df.Column_New=='Total Year-Round Beds')&(final_df.Program_Type=='ES, TH, SH')]
changes_df.insert(2, '2018 to 2019', (changes_df.loc[:,2019]-changes_df.loc[:,2018])/changes_df.loc[:,2018]*100)
changes_df.insert(4, '2019 to 2020', (changes_df.loc[:,2020]-changes_df.loc[:,2019])/changes_df.loc[:,2019]*100)
changes_df.insert(6, '2020 to 2021', (changes_df.loc[:,2021]-changes_df.loc[:,2020])/changes_df.loc[:,2020]*100)
changes_df.insert(8, '2021 to 2022', (changes_df.loc[:,2022]-changes_df.loc[:,2021])/changes_df.loc[:,2021]*100)
changes_df.sort_values(['2021 to 2022'], ascending=False)
```

```python
changes_df.to_csv('HIC_GAvsUSA_changes.csv', index=False)
```
